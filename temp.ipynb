```python
def PSNR(super_resolution, high_resolution):
    """Compute the peak signal-to-noise ratio, measures quality of image."""
    # Max value of pixel is 255
    psnr_value = tf.image.psnr(high_resolution, super_resolution, max_val=255)[0]
    return psnr_value
```

# Tranfer learning

```python
from model import edsr
```

```python
EDSR_x2_l1 = tf.keras.models.load_model("/content/drive/MyDrive/mah_ws/ai_prj/prj_ImageSuper-Resolution/save_model/edsr/shuffle/200_baseline_x2_l1_32.h5",custom_objects={'PSNR':PSNR})
EDSR_x2_l1.load_weights("/content/drive/MyDrive/mah_ws/ai_prj/prj_ImageSuper-Resolution/checkpoint/edsr/shuffle/200_baseline_x2_l1_32.h5")
```

```python
for i in range(len(EDSR_x2_l1.layers)):
  EDSR_x2_l1.layers[i].trainable = False 
```

```python
outp = EDSR_x2_l1.layers[-5].output
outp = edsr.upsample(outp,4,128)
outp = tf.keras.layers.Conv2D(3, 3, padding='same',name='conv2d_1_tranfer')(outp)

outp = tf.keras.layers.Lambda(edsr.denormalize,name='Denormarlize')(outp)
```

```python
EDSR_x4_l1 = tf.keras.models.Model(inputs = EDSR_x2_l1.input, outputs= outp, name = 'Tranfer_learning')
EDSR_x4_l1.summary()
```

```python
from model import srcnn

inp = tf.keras.layers.Input(shape = (None,None,3))
out = srcnn.upsample(inp, 4, 64)
for i in range(4,7):
  l = SRCNN_x2_l1.layers[i]
  out = l(out)
```

```python
SRCNN_x4_l1 = tf.keras.models.Model(inputs = inp, outputs = out)
SRCNN_x4_l1.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint, ReduceLROnPlateau


ds_train = tfds.load(f'div2k/bicubic_x4',split='train[:75%]',as_supervised=True)
ds_valid = tfds.load(f'div2k/bicubic_x4',split='train[75%:]',as_supervised=True)
ds_test = tfds.load(f'div2k/bicubic_x4',split='validation',as_supervised=True)

train = agumentation.preprocessing(ds_train, scale = 4)
valid = agumentation.preprocessing(ds_valid, scale = 4,batch_size=1,repeat_count=1)
test = agumentation.preprocessing(ds_test,batch_size=1,repeat_count=1,random_transform=False)
```

```python
optim = Adam(lr=3e-4)
with tf.device('/gpu:0'):
  EDSR_x4_l1.compile(optimizer= optim , loss = 'mae', metrics=[PSNR])
```

```python
name_exp = 'tranfer/EDSR_x4_l1'

reducelr = ReduceLROnPlateau( monitor='val_loss',factor=tf.math.exp(-0.1), patience=8, verbose=1, mode='auto', min_lr=1e-5)
checkpoint = ModelCheckpoint(f"{root_dir}/checkpoint/srcnn/{name_exp}.h5", monitor="val_loss", verbose=0, save_best_only=True,save_weight_only = True, mode="auto",)
# print(len(train)," : ",len(valid))
EDSR_x4_l1.fit(train,validation_data=valid ,epochs=100,verbose=True, callbacks=[reducelr,checkpoint],steps_per_epoch=100)
```

```python
loss = pd.DataFrame(EDSR_x4_l1.history.history)
    
loss[['val_loss']].plot()
loss[['val_PSNR']].plot()

## save hist
with open(f'{root_dir}/save_model/edsr/{name_exp}.json', mode='wt') as f:
    loss.to_json(f)


EDSR_x4_l1.save(f'{root_dir}/save_model/edsr/{name_exp}.h5')
# SRCNN.summary()
EDSR_x4_l1.evaluate(test)
```
